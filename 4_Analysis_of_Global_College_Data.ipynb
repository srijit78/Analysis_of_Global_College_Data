# Visualizing Gender Distribution per **Country**

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Country', y='Male', data=df, label='Male', color='skyblue')
sns.barplot(x='Country', y='Female', data=df, label='Female', color='lightcoral', alpha=0.8)
plt.xticks(rotation=45)
plt.title("Gender Distribution per Country")
plt.legend()
plt.show()
```

# Branch-wise Academic Performance (Average CGPA by Branch)**bold text**

```python
# Calculate the average CGPA per branch
branch_avg = df.groupby('Branch')['CGPA'].mean().reset_index()

# Sort values for cleaner plotting
branch_avg = branch_avg.sort_values(by='CGPA', ascending=False)

# Display the averages
print(branch_avg)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Branch', y='CGPA', data=branch_avg, palette='viridis')
plt.xticks(rotation=45)
plt.title("Average CGPA across Different Branches")
plt.ylabel("Average CGPA")
plt.xlabel("Branch")

# Add the values on top of each bar for clarity
for index, value in enumerate(branch_avg['CGPA']):
    plt.text(index, value + 0.05, f"{value:.2f}", ha='center', fontsize=10)

plt.show()
```

```python
# Highlight highest and lowest branches
#The highest CGPA branch (green)
#The lowest CGPA branch (red)
#The rest (blue)

colors = ['green' if x == branch_avg['CGPA'].max() else
          'red' if x == branch_avg['CGPA'].min() else 'skyblue'
          for x in branch_avg['CGPA']]

plt.figure(figsize=(10,6))
sns.barplot(x='Branch', y='CGPA', data=branch_avg, palette=colors)
plt.xticks(rotation=45)
plt.title("Average CGPA by Branch (Highest & Lowest Highlighted)")
plt.ylabel("Average CGPA")
plt.xlabel("Branch")

for index, value in enumerate(branch_avg['CGPA']):
    plt.text(index, value + 0.05, f"{value:.2f}", ha='center', fontsize=10)

plt.show()
```

# Country vs Placement **Rate**

```python
# Calculate average placement rate for each country
country_avg = df.groupby('Country')['Placement Rate'].mean().reset_index()

# Sort by placement rate (optional for better visualization)
country_avg = country_avg.sort_values(by='Placement Rate', ascending=False)

# View the results
print(country_avg)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Country', y='Placement Rate', data=country_avg, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Average Placement Rate by Country")
plt.ylabel("Average Placement Rate (%)")
plt.xlabel("Country")

# Add text labels on top of bars
for index, value in enumerate(country_avg['Placement Rate']):
    plt.text(index, value + 0.5, f"{value:.2f}", ha='center', fontsize=10)

plt.show()
```

```python
# Create color map highlighting extremes
colors = ['green' if x == country_avg['Placement Rate'].max() else
          'red' if x == country_avg['Placement Rate'].min() else
          'skyblue' for x in country_avg['Placement Rate']]

plt.figure(figsize=(10,6))
sns.barplot(x='Country', y='Placement Rate', data=country_avg, palette=colors)
plt.xticks(rotation=45)
plt.title("Average Placement Rate by Country (Highest & Lowest Highlighted)")
plt.ylabel("Placement Rate (%)")
plt.xlabel("Country")

# Add text labels
for index, value in enumerate(country_avg['Placement Rate']):
    plt.text(index, value + 0.5, f"{value:.2f}", ha='center', fontsize=10)

plt.show()
```

# Faculty Count vs Research **Output**

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df['Faculty Count'].describe())
print(df['Research Papers Published'].describe())
```

```python
# Group data by branch and compute averages
faculty_research = df.groupby('Branch')[['Faculty Count', 'Research Papers Published']].mean().reset_index()

# View data
print(faculty_research)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Faculty Count',
                y='Research Papers Published',
                hue='Branch',
                data=faculty_research,
                s=100)

plt.title("Faculty Count vs Research Output (Branch-wise Average)")
plt.xlabel("Average Faculty Count")
plt.ylabel("Average Research Papers Published")

# Annotate each point with its branch name
for i in range(faculty_research.shape[0]):
    plt.text(faculty_research['Faculty Count'][i]+1,
             faculty_research['Research Papers Published'][i],
             faculty_research['Branch'][i], fontsize=9)

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='Faculty Count',
            y='Research Papers Published',
            data=faculty_research,
            scatter_kws={'color':'skyblue','s':80},
            line_kws={'color':'red'})
plt.title("Faculty Count vs Research Output with Trend Line")
plt.xlabel("Average Faculty Count")
plt.ylabel("Average Research Papers Published")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Faculty Count', y='Research Papers Published', hue='Branch', data=df)
plt.title("Faculty Count vs Research Output")
plt.xlabel("Faculty Count")
plt.ylabel("Research Papers Published")
plt.show()
```
